```python
import shutil

quantity = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']
view = ['side', 'top']
order = ['00001.JPG', '00002.JPG', '00003.JPG', '00004.JPG', '00005.JPG', '00006.JPG', '00007.JPG', '00008.JPG', '00009.JPG', '00010.JPG']
sub = ['00011.JPG', '00012.JPG', '00013.JPG', '00014.JPG', '00015.JPG', '00016.JPG', '00017.JPG', '00018.JPG', '00019.JPG', '00020.JPG']
save_path = ['/content/gdrive/MyDrive/DCD_2022/음식 이미지 및 영양정보 텍스트/train_sampling']
error = []

filenames = os.listdir(data_dir)
path = [data_dir]
for i in filenames:
  save_path2 = save_path[:]
  save_path2.append(i)
  save_path2 = os.path.join(*save_path2)
  folder_path = path[:]
  folder_path.append(i)
  file_path = os.path.join(*folder_path)
  file_name = os.listdir(file_path)
  file_name = file_name[0].split('_')
  for v in view:
    file_name[0] = v
    for q in quantity:
      file_name[4] = q
      for o in order:
        try:
          file_name[5] = o
        except:
          error.append(i)
          break
        file = '_'.join(file_name)
        folder_path2 = folder_path[:]
        folder_path2.append(file)
        file_path = os.path.join(*folder_path2)
        sub2 = sub[:]
        while os.path.isfile(file_path) == False:
          file_name[5] = sub2[0]
          file = '_'.join(file_name)
          del sub2[0]
          folder_path2 = folder_path[:]
          folder_path2.append(file)
          file_path = os.path.join(*folder_path2)
        shutil.copy(file_path, save_path2)
```
